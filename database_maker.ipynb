```python
import faker
import random
import pandas as pd

fake = faker.Faker()
```

```python
def generate_users(num_users=10):
    users = []
    for i in range(num_users):
        user = {
            "username": fake.user_name(),
            "password": fake.password(),
            "email": fake.email(),
            "name": fake.name(),
        }
        users.append(user)
    return users
```

```python
def generate_fake_posts(num_posts=20):
    data = []
    for i in range(num_posts):
        post = {
            "id": i + 1,
            "title": fake.sentence(nb_words=6),
            "content": fake.paragraph(nb_sentences=25),
            "author": fake.user_name(),
            "created_at": fake.date_time_this_year().isoformat(),
            "views": random.randint(0, 10000),
            "is_published": random.choice([True, False])
        }
        data.append(post)
    return data
```

```python
users = generate_users(50)
posts = generate_fake_posts(100)
```

```python
user_df = pd.DataFrame(columns =  ['username', 'password', 'email', 'name'])
for user in users:
    user_df.loc[len(user_df)] = user
```

```python
posts_df = pd.DataFrame(columns = ['id', 'title', 'content', 'author', 'created_at', 'views', 'is_published'])
for post in posts:
    posts_df.loc[len(posts_df)] = post
```

```python
user_df.to_csv('fake_data/users.csv',  index=False)
posts_df.to_csv("fake_data/posts.csv", index=False)
```
